```python
# data science with python
# applications of data science
# step1 : import nessary libraries
# step3 : data cleaning and preprocessing.
# step4: EDA Exploratory Data Analysis
# step5: Distribution students with age and whether they have girlfriend or boyfriend.
# Gender who have girlfriend or boyfriend.
# Do you eat, age,gender and girlfriend or boyfriend.
# living allowance and girlfriend or boyfriend.
# Those with girlfriend or boyfriend and siblings
# correlation Heatmaps , correlation matrix and plot the heatmaps.
# Decission making , insights from the data  and data analysis conclusion.
#1. age and girlfriend or boyfriend
#2: Gender and girlfriend or boyfriend.
#3: correction analysis.
import os
import sys
print(sys.executable)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Check current directory and adjust if necessary
print("Current Directory:", os.getcwd())

# step 2: load the dataset
# Replace 'error_bad_lines' with 'on_bad_lines' and set it to 'skip'
df = pd.read_csv('niceguys.csv', on_bad_lines='skip')  
print(df.head(5))

# Fixing column names
df.columns = ['number_siblings', 'name', 'sex', 'age', 'living_allowance', 'occupation', 'meals_per_day', 'has_partner']

# Convert living_allowance to a numerical field, removing commas
df['living_allowance'] = df['living_allowance'].replace(',', '', regex=True).astype(float)

# Handle missing values, for simplicity, fill them with the median or mode
df['age'] = df['age'].fillna(df['age'].median())
df['living_allowance'] = df['living_allowance'].fillna(df['living_allowance'].median())
df['has_partner'] = df['has_partner'].fillna(df['has_partner'].mode()[0])

print(df.head())

# Distribution of age and partner status
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='has_partner', multiple='stack')
plt.title('Distribution of Age and Relationship Status')
plt.show()

# Gender and partner status
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='sex', hue='has_partner')
plt.title('Gender vs Relationship Status')
plt.show()

# Exploring correlation between numerical data
correlation = df[['age', 'living_allowance', 'number_siblings']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Calculate the percentage of individuals with a partner per gender
partner_by_gender = df.groupby('sex')['has_partner'].mean() * 100
print(partner_by_gender)
```
